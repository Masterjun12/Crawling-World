```python
import pandas as pd
import numpy as np
import math

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver import ActionChains as AC
from selenium.webdriver.common.keys import Keys
from tqdm import tqdm
from tqdm import tqdm_notebook
import re
from time import sleep
import time
```

```python
# 플레이스토어 크롬창 띄우기
driver = webdriver.Chrome("chromedriver.exe") # 크롬 드라이버 로드
driver.get("https://play.google.com/store/search?q=%EA%B8%B0%EC%95%84%EC%9E%90%EB%8F%99%EC%B0%A8&c=apps")   # 네이버카페 띄우기
time.sleep(1)
```

```python
# title 크롤링 시작
title_list=[]
overlays = ".b8cIId.ReQCgd.Q9MA7b"                                 
tit = driver.find_elements_by_css_selector(overlays)         # title
for t in tit:
    title = t.get_attribute('title')
    title_list.append(title)

title_list
```

```python
# company 크롤링 시작
company_list=[]
overlays = ".b8cIId.ReQCgd.KoLSrc"                                 
comp = driver.find_elements_by_css_selector(overlays)         # company
for c in comp:
    company = c.text
    company_list.append(company)

company_list
```

```python
tmp = pd.DataFrame({'title' : title_list, 'company' : company_list})  
print(len(tmp))
tmp
```

```python
tmp.to_csv("data/kiacar({0}).csv".format(len(tmp)))
```

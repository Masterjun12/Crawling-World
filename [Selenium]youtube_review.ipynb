```python
# 크롬창 띄우기
from selenium import webdriver
driver = webdriver.Chrome("chromedriver.exe")  # 크롬 드라이버 로드

driver.get("https://www.youtube.com/watch?v=SfZWFDs0LxA")  # 웹 페이지로 이동
```

```python
# 스크롤 다운
import time
def scroll_down(driver):
  driver.execute_script("window.scrollTo(0,9999999999999)")   
  time.sleep(2)
    
i = 0
while i < 10:
  scroll_down(driver) # 스크롤 다운
  i += 1
```

## 리뷰 크롤링

```python
# selector:  #id와 .class
things = driver.find_elements_by_css_selector("#content-text")    
```

```python
## import requests
## bs4 import BeautifulSoup

import requests
from bs4 import BeautifulSoup

## F12 -> Network -> XHR -> 새로고침 -> Header가서 POST면 맽 밑에 Form data 넣어주기
## url / param / requests.post() / .text / BS
url = 'http://fifaonline4.nexon.com/datacenter/PlayerList'
param = {
    'n1Strong': '1',
    'n1Grow': '0',
    'strSkill1': 'liveperformance1',
    'strSkill2': 'liveperformance2',
    'strSkill3': 'liveperformance3',
    'strSkill4': 'liveperformance4',
    'strSearchStatus': 'off',
    'strOrderby': '',
    'n1LivePerfomance': '1',
    'n4PageNo': '1rd=0.7256169577732083'
}

r = requests.post(url, param)
html = r.text

soup = BeautifulSoup(html, 'lxml')   # lxml : bs 쓸 때 기본. 
```

```python
## soup.find_all('div', {'class':'player_info'})
target_peoples = soup.find_all('div', {'class':'name'})  #div 태그에 class 중 name 크롤링
target_peoples
```

```python
target_info_all = {}

index_stats = ['급여','이번주 퍼포먼스','지난주 퍼포먼스','2주전 퍼포먼스','3주전 퍼포먼스']
index_ovrs = ['OVR','기본 OVR','퍼포먼스 변화량']

for people in target_peoples:
    target_ovrs = people.parent.parent.parent.parent.find_all('div', {'class':'td td_ar ovr'})
    target_stats = people.parent.parent.parent.parent.find_all('div', {'class':'td td_ar'})

    target_info = {}
    for i,stat in enumerate(target_stats):
    #     print(index_stats[i], stat.text.strip())
    #     target_info[index_stats[i]] = stat.text.strip()
        if stat.text.strip() == 'ㅡ':
            target_info[index_stats[i]] = 0
        else:
            target_info[index_stats[i]] = float(stat.text.strip())

    for i, ovr in enumerate(target_ovrs[0].text.split()):
    #     print(index_ovrs[i], ovr)
    #     target_info[index_ovrs[i]] = float(ovr)
        if ovr[0] == '(':
            target_info[index_ovrs[i]] = -1 * float(ovr.replace('(','').replace(')',''))
        else:
            target_info[index_ovrs[i]] = float(ovr)
    
    # Customized Statistics for Variation(%)

    if target_info['이번주 퍼포먼스'] == 0 or target_info['지난주 퍼포먼스'] == 0:
        target_info['지난주 대비 변화량'] = '계산 불가'
    else:
        target_info['지난주 대비 변화량'] = 
        (float(target_info['이번주 퍼포먼스'])/float(target_info['지난주 퍼포먼스'])-1)
        
    target_info_all[people.text] = target_info
```

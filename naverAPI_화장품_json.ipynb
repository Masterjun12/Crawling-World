```python
# 네이버 카페글 "화장품" 크롤링
# json 으로 크롤링

# 문제 : 네이버 카페글 정보 추출, csv로 저장해서 csv 파일 제출
# 제출 데이터 : title / link / description / cafename / cafeurl

# 네이버 데이터랩 - 카페글 -> api 활용해서 크롤링해보기
# api 필요정보 : https://developers.naver.com/docs/search/cafearticle/
```

```python
import requests
import os
import pandas
import time
import re
from w3lib.html import replace_entities
```

```python
# 요청할 페이지의 주소
site = 'https://openapi.naver.com/v1/search/cafearticle.json'

# 헤더정보
client_id = 'QXuk4f_FKXswmmjcYyd8'
client_secret = 'cOo1RIqRYo'

# 파라미터
query = '화장품'

header = {
    'X-Naver-Client-Id' : client_id,
    'X-Naver-Client-Secret' : client_secret
}
param = {
    'query' : query
}

# 요청한다.
response = requests.get(site, headers=header, params=param)

# 응답결과를 json 객체로 만들어준다.
root = response.json()

# 응답결과를 json 객체로 만들어준다.
root = response.json()

# 최대 글 개수를 가져온다.
total = int(root['total'])
start = int(root['start'])
display = int(root['display'])

# <>태그들을 지워주는 함수 생성
def remove_tag(content):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', content)
    text = replace_entities(cleantext)
    return text

while True :           # 무한 반복
    time.sleep(1)      # 1초의 텀을 가지고 크롤링

    # 리스트의 수만큼 반복한다.
    for item in root['items'] :
        title = remove_tag(item['title'])
        link = remove_tag(item['link'])
        description = remove_tag(item['description'])
        description = re.sub('&apos;', "'", description)
        cafename = remove_tag(item['cafename'])
        cafeurl = remove_tag(item['cafeurl'])

        df = pandas.DataFrame([[title, link, description, cafename, cafeurl]])

        if os.path.exists('naverAPI_화장품_json.csv') == False :
            df.columns = ['title', 'link', 'description', 'cafename', 'cafeurl']
            df.to_csv('naverAPI_화장품_json.csv', index=False, encoding='utf-8-sig')
        else:
            df.to_csv('naverAPI_화장품_json.csv', index=False, encoding='utf-8-sig', mode='a', header=False)

    print('start :', start)

    if start < 1000 and start < total :
        start += display
        if start > 1000 :
            start = 1000

        param['start'] = start
    else :
        break
```
